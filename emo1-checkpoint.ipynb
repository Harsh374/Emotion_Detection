**Preparing data for training**


```python
img_size = 48
batch_size = 64
from tensorflow.keras.preprocessing.image import ImageDataGenerator
datagen_train = ImageDataGenerator()
train_generator = datagen_train.flow_from_directory("C:/Users/harsh/Desktop/Emotion_detection_course/archive/train",
                                                    target_size = (img_size,img_size),
                                                    color_mode ="grayscale",
                                                    batch_size = batch_size,
                                                    class_mode = "categorical",
                                                    shuffle = True
                                                    )
datagen_validation = ImageDataGenerator()
validation_generator = datagen_train.flow_from_directory( "C:/Users/harsh/Desktop/Emotion_detection_course/archive/train",
                                                    target_size = (img_size,img_size),
                                                    color_mode ="grayscale",
                                                    batch_size = batch_size,
                                                    class_mode = "categorical",
                                                    shuffle = True
                                                    )

```

**Defining Model**


```python
def Convolution(input_tensor, filters, kernel_size):
    x = Conv2D(filters = filters, kernel_size = kernel_size, padding= "same")(input_tensor)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D(pool_size = (2,2))(x)
    x = Dropout(0.25)(x)
     
    return x

def Dense_f(input_tensor, nodes):
    x = Dense(nodes)(input_tensor)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Dropout(0.25)(x)
    
    return x

def model_fer(input_shape):
    inputs = Input(input_shape)
    conv_1 = Convolution(inputs,32,(3,3))
    conv_2 = Convolution(inputs,64,(5,5))
    conv_3 = Convolution(inputs,120,(3,3))
    
    flatten = Flatten()(conv_3)
    
    dense_1 = Dense_f(flatten,256)
    
    output = Dense(7,activation = "softmax")(dense_1)
    model = Model(inputs = [inputs], outputs = [output])
    
    model.compile(loss = ['categorical_crossentropy'], optimizer = 'adam', metrics = ['accuracy'])
    
    return model
model = model_fer((48,48,1))
model.summary()




 
```

**Initializing the model**


```python
epochs = 15
steps_per_epoch = train_generator.n//train_generator.batch_size
validation_steps = validation_generator.n//validation_generator.batch_size

checkpoint = ModelCheckpoint("model.weights.h5", monitor = 'val_accuracy',save_weights_only = True , mode = 'max',verbose = 1)
callback = [checkpoint]
```

**Training the model**


**...The next cell takes time to complete**

```python
epochs = 15
steps_per_epoch = train_generator.n//train_generator.batch_size
validation_steps = validation_generator.n//validation_generator.batch_size

checkpoint = ModelCheckpoint("model.weights.h5", monitor = 'val_accuracy',save_weights_only = True , mode = 'max',verbose = 1)
callback = [checkpoint]

history = model.fit(
x = train_generator,
steps_per_epoch = steps_per_epoch,
epochs = epochs,
validation_data = validation_generator,
validation_steps = validation_steps,
callbacks = callback)
```

**Evaluate the model**

```python
model.evaluate(validation_generator)
```
